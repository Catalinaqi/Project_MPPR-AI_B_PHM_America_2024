```python
import sklearn as sk
import pandas as pd
```

### caricamento dati

```python
x = pd.read_csv('../X_train.csv')
print(x)
y = pd.read_csv('../y_train.csv')
print(y)
full = pd.merge(x,y,on='id')
print(full)
```

### generazione feature

```python
full['trq_target'] = (full['trq_measured']*100)/(full['trq_margin']+100)
```

### suddivisione dati

```python
X_features = [c for c in x.columns if c != 'id']
y_features = [c for c in y.columns if c != 'id']
X_train, X_nt, y_train, y_nt = sk.model_selection.train_test_split(full[X_features], full[y_features+['trq_target']], test_size=0.4)
X_test, X_val, y_test, y_val = sk.model_selection.train_test_split(X_nt, y_nt, test_size=0.5)
```

### addestramento modello regressione

```python
reg = sk.linear_model.LinearRegression()
reg.fit(X_train, y_train['trq_target'])
print(f'coefficienti: {reg.coef_}')
print(f'f0: {reg.intercept_}')
```

### valutazione modello regressione

```python
trq_target_guess = reg.predict(X_test)
guess = 100*((X_test['trq_measured']-trq_target_guess)/trq_target_guess)
error_reg = sk.metrics.mean_squared_error(guess, y_test['trq_margin'])
print(f'error: {error_reg}')
```
