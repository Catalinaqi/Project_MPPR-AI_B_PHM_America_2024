```python
from pathlib import Path

def find_project_root(start: Path) -> Path:
    p = start.resolve()
    for _ in range(10):
        if (p / "config").exists() and (p / "src").exists():
            return p
        p = p.parent
    raise RuntimeError(f"Project root not found from: {start}")

PROJECT_ROOT = find_project_root(Path.cwd())
print("PROJECT_ROOT =", PROJECT_ROOT)

# IMPORTANT: work with the SAME dict object
variables = ctx.variables

# output_root ABSOLUTO
ctx.output_root = str(PROJECT_ROOT / "out")

# make dataset paths ABSOLUTE
for k in ["x_train_path", "y_train_path", "x_test_path", "x_validation_path"]:
    if k in variables:
        p = Path(variables[k])
        if not p.is_absolute():
            variables[k] = str((PROJECT_ROOT / p).resolve())

# HARD check: must exist
for k in ["x_train_path", "y_train_path"]:
    print(k, "=>", variables[k], "| exists:", Path(variables[k]).exists())
```
