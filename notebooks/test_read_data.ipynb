```python
# --- Dataset summary + split percentages (English comments) ---
import pandas as pd

# Update these paths if needed
X_train_path = x_path
y_train_path = y_path
X_test_path  = x_test_path
X_val_path   = x_validation_path

# Read CSVs
X_train = pd.read_csv(X_train_path)
y_train = pd.read_csv(y_train_path)
X_test  = pd.read_csv(X_test_path)
X_val   = pd.read_csv(X_val_path)

# --- 1) Summary table: rows & columns ---
summary = pd.DataFrame([
    {"Dataset": "Train (X)",       "File": "X_train.csv",       "Rows": X_train.shape[0], "Columns": X_train.shape[1]},
    {"Dataset": "Train (y)",       "File": "Y_train.csv",       "Rows": y_train.shape[0], "Columns": y_train.shape[1]},
    {"Dataset": "Validation (X)",  "File": "X_validation.csv",  "Rows": X_val.shape[0],   "Columns": X_val.shape[1]},
    {"Dataset": "Test (X)",        "File": "X_test.csv",        "Rows": X_test.shape[0],  "Columns": X_test.shape[1]},
])

print("=== Dataset Summary (Rows & Columns) ===")
display(summary)

# Sanity check: X_train and y_train must have the same number of rows
if X_train.shape[0] != y_train.shape[0]:
    print(f"⚠️ Warning: X_train rows ({X_train.shape[0]}) != y_train rows ({y_train.shape[0]})")

# --- 2) Split percentages (based on number of samples) ---
# Note: Train split is counted using X_train rows. y_train is the label file for the same samples.
counts = {
    "Train": X_train.shape[0],
    "Validation": X_val.shape[0],
    "Test": X_test.shape[0],
}
total = sum(counts.values())

percentages = {k: (v / total) * 100 for k, v in counts.items()}

split_table = pd.DataFrame([
    {"Split": k, "Rows": v, "Percentage (%)": round(percentages[k], 2)}
    for k, v in counts.items()
])

print("=== Split Percentages (by rows) ===")
display(split_table)

# Optional: quick print
print("Total samples:", total)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

paths = {
    "Train (X_train)": str(x_path),
    "Test (X_test)": str(x_test_path),
    "Validation (X_validation)": str(x_validation_path),
}

# contar filas
counts = {name: pd.read_csv(p).shape[0] for name, p in paths.items()}
total = sum(counts.values())
perc = {k: v/total*100 for k, v in counts.items()}

print("Filas por split:", counts)
print("Porcentajes (%):", {k: round(v,3) for k,v in perc.items()})

# Pie chart
plt.figure()
plt.pie(counts.values(), labels=[f"{k}\n{perc[k]:.1f}%" for k in counts], autopct="")
plt.title("Distribución de filas: Train vs Test vs Validation")
plt.show()

# Bar chart
plt.figure()
plt.bar(list(counts.keys()), list(counts.values()))
plt.title("Nº de filas por split")
plt.ylabel("Filas")
plt.xticks(rotation=15, ha="right")
plt.show()
```

```python
import pandas as pd

# 1) Read CSVs (with consistent parameters)


X_train_path = x_path
y_train_path = y_path


X = pd.read_csv(X_train_path, sep=",", encoding="utf-8", decimal=".", low_memory=False)
Y = pd.read_csv(y_train_path, sep=",", encoding="utf-8", decimal=".", low_memory=False)

# 2) The id column should be int for merging, ensure correct type
X["id"] = X["id"].astype(int)
Y["id"] = Y["id"].astype(int)

# 3) Consistency checks: check for duplicate ids in both X and Y
print("X dup id:", X["id"].duplicated().sum())
print("Y dup id:", Y["id"].duplicated().sum())

# 4) Merge on "id" with inner join and validate one-to-one relationship
df_train = X.merge(Y, on="id", how="inner", validate="one_to_one")

print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("Merged shape:", df_train.shape)

# 5) Display first 10 rows of the merged DataFrame
display(df_train.head(10))
```

```python
# df_train = ... (il tuo merge)

rename_map = {
    # X (feature)
    "trq_measured": "TORQUE_MEASURED",
    "oat": "OAT",
    "mgt": "MGT",
    "pa": "PA",
    "ias": "IAS",
    "np": "NP",
    "ng": "NG",

    # Y (target)
    "faulty": "FAULTY",
    "trq_margin": "TRQ_MARGIN",
}

df_train_named = df_train.rename(columns=rename_map)

print("Columns (renamed):")
print(df_train_named.columns.tolist())

display(df_train_named.head(10))
```
